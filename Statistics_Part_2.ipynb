# **üìò Theoretical Part (Hypothesis Testing & Statistical Concepts)**


1. What is hypothesis testing in statistics?

  Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data. It involves:

* Formulating two hypotheses:

  * **Null Hypothesis (H‚ÇÄ):** Assumes no effect or no difference.
  * **Alternative Hypothesis (H‚ÇÅ):** Indicates the presence of an effect or difference.

* Using sample data to determine whether to reject H‚ÇÄ in favor of H‚ÇÅ.

---

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

* **Null Hypothesis (H‚ÇÄ):** A statement of no effect (e.g., ‚ÄúThe mean weight of apples is 200g.‚Äù).
* **Alternative Hypothesis (H‚ÇÅ or Ha):** A statement that contradicts the null (e.g., ‚ÄúThe mean weight is not 200g.‚Äù).
* Hypothesis tests aim to gather enough evidence against H‚ÇÄ.

---

3. What is the significance level in hypothesis testing, and why is it important?

* The **significance level (Œ±)** is the probability of rejecting the null hypothesis when it is actually true (**Type I error**).
* Common values: **0.05**, **0.01**, **0.10**.
* Lower Œ± ‚Üí stricter decision threshold.

---

4. What does a P-value represent in hypothesis testing?

* The **P-value** is the probability of obtaining a test statistic at least as extreme as the one observed, under the assumption that H‚ÇÄ is true.

---

5. How do you interpret the P-value in hypothesis testing?

* **P < Œ±**: Reject H‚ÇÄ (significant result).
* **P ‚â• Œ±**: Fail to reject H‚ÇÄ (not significant).
* A **small P-value** suggests strong evidence against H‚ÇÄ.

---

6. What are Type 1 and Type 2 errors in hypothesis testing?

| Type        | Error Description                                     |
| ----------- | ----------------------------------------------------- |
| Type I (Œ±)  | Rejecting H‚ÇÄ when it's true (False positive)          |
| Type II (Œ≤) | Failing to reject H‚ÇÄ when it's false (False negative) |

---

7. What is the difference between a one-tailed and a two-tailed test  in hypothesis testing?

* **One-tailed test**: Tests deviation in one direction (e.g., greater than).
* **Two-tailed test**: Tests deviation in both directions (e.g., not equal).

---

8. What is the Z-test, and when is it used in hypothesis testing?

* A **Z-test** compares a sample mean to a known population mean when population standard deviation is known.
* Used when **n > 30** (large sample) and population is normally distributed.

---

9. How do you calculate the Z-score, and what does it represent  in hypothesis testing?

$$
Z = \frac{(X - \mu)}{\sigma}
$$

* Represents how many standard deviations a value **X** is from the mean **Œº**.
* Useful for standardizing data.

---

10. What is the T-distribution, and when should it be used instead of the normal distribution?

* The **T-distribution** is used when:

  * Sample size is **small (n < 30)**.
  * Population standard deviation is **unknown**.
* T-distribution is wider and more variable than normal distribution.

---

11. What is the difference between a Z-test and a T-test?

| Test   | Use Case                           | Known œÉ | Sample Size    |
| ------ | ---------------------------------- | ------- | -------------- |
| Z-test | Known population std. dev (œÉ)      | Yes     | Large (n > 30) |
| T-test | Unknown œÉ and/or small sample size | No      | Small (n < 30) |

---
12. What is the T-test, and how is it used in hypothesis testing?

* **T-test** checks if the means of two groups are statistically different.
* Types:

  * **One-sample**
  * **Independent two-sample**
  * **Paired sample**

---

13. What is the relationship between Z-test and T-test in hypothesis testing?

* Both test means but differ in assumptions:

  * Z-test: known œÉ
  * T-test: unknown œÉ
* T-test approximates Z-test as sample size increases.

---

14. What is a confidence interval, and how is it used to interpret statistical results?

* A **confidence interval (CI)** gives a range of plausible values for a parameter.
* E.g., 95% CI = we‚Äôre 95% confident the true mean lies in that range.

---

15. What is the margin of error, and how does it affect the confidence interval?

* **Margin of Error (MoE)** is the maximum expected difference between the sample estimate and the population parameter.
* CI = Estimate ¬± MoE

---

16. How is Bayes' Theorem used in statistics, and what is its significance?

Bayes' Theorem:

$$
P(A|B) = \frac{P(B|A)P(A)}{P(B)}
$$

* Updates the probability of a hypothesis based on new evidence.
* Used in:

  * Spam filters
  * Medical diagnostics
  * Machine learning

---

17. What is the Chi-square distribution, and when is it used?

* A **Chi-square (œá¬≤) distribution** is used for **categorical data**.
* Applications:

  * Goodness-of-fit test
  * Test of independence

---

18. What is the Chi-square goodness of fit test, and how is it applied?

* Tests whether observed categorical frequencies match expected frequencies.
* Formula:

$$
\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}
$$

---

19. What is the F-distribution, and when is it used in hypothesis testing?

* The **F-distribution** compares two variances.
* Used in:

  * ANOVA (analysis of variance)
  * Regression models

---

20. What is an ANOVA test, and what are its assumptions?

* **ANOVA** tests if three or more group means are equal.
* Assumptions:

  * Independence
  * Normality
  * Equal variances

---

21. What are the different types of ANOVA tests?

| Type              | Description                     |
| ----------------- | ------------------------------- |
| One-way           | One independent variable        |
| Two-way           | Two independent variables       |
| Repeated measures | Same subjects tested repeatedly |

---

22. What is the F-test, and how does it relate to hypothesis testing?

* An **F-test** compares variances of two populations.
* If F-ratio is large, variances differ significantly.
* It‚Äôs the basis of ANOVA.

15.  Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

# Given data
sample = np.array([90, 110, 95, 105, 100, 98, 102, 97, 99, 103])  # Sample sales data
population_mean = 100  # Claimed mean
population_std = 10     # Known standard deviation
sample_size = len(sample)

# Step 1: Compute Sample Mean
sample_mean = np.mean(sample)

# Step 2: Compute Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Step 3: Compute P-value
p_value = 2 * (1 - norm.cdf(abs(z_score)))  # Two-tailed test

# Step 4: Conclusion
alpha = 0.05  # Significance level
if p_value < alpha:
    print(f"Reject the null hypothesis (H0). Z-score: {z_score:.2f}, P-value: {p_value:.4f}")
else:
    print(f"Fail to reject the null hypothesis (H0). Z-score: {z_score:.2f}, P-value: {p_value:.4f}")
```

16. Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
import scipy.stats as stats

# Step 1: Generate sample data
data = np.random.normal(loc=100, scale=15, size=30)  # Mean=100, Std=15, Sample Size=30

# Step 2: Compute sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Use ddof=1 for sample standard deviation

# Step 3: Set confidence level (95%) and find Z-score
confidence_level = 0.95
z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Step 4: Compute margin of error
margin_of_error = z_score * (sample_std / np.sqrt(len(data)))

# Step 5: Calculate confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Display results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

24.  Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
distribution.

```python
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(size=1000, mean=0, std=1):
    # Generate normally distributed data
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Calculate skewness and kurtosis
    skewness = skew(data)
    kurt = kurtosis(data)  # Fisher's definition (normal = 0)

    # Print the results
    print("Skewness:", skewness)
    print("Kurtosis:", kurt)

# Call the function
calculate_skewness_kurtosis()
```

1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results.

```python
import numpy as np
from scipy.stats import norm

# Sample dataset (daily sales)
sample = np.array([90, 110, 95, 105, 100, 98, 102, 97, 99, 103])  # 10 data points
population_mean = 100  # Claimed mean
population_std = 10     # Known standard deviation
sample_size = len(sample)

# Compute Sample Mean
sample_mean = np.mean(sample)

# Compute Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Compute P-value
p_value = 2 * (1 - norm.cdf(abs(z_score)))  # Two-tailed test

# Define significance level
alpha = 0.05

# Interpretation
if p_value < alpha:
    print(f"Reject the null hypothesis (H‚ÇÄ). Z-score: {z_score:.2f}, P-value: {p_value:.4f}")
else:
    print(f"Fail to reject the null hypothesis (H‚ÇÄ). Z-score: {z_score:.2f}, P-value: {p_value:.4f}")
```

4.  Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Sample Data
sample = np.array([92, 108, 95, 107, 101, 99, 105, 97, 100, 102])  # Sample values
population_mean = 100  # Claimed mean
population_std = 10     # Known standard deviation
sample_size = len(sample)

# Step 2: Compute Sample Mean
sample_mean = np.mean(sample)

# Step 3: Compute Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Step 4: Compute P-value
p_value = 2 * (1 - norm.cdf(abs(z_score)))  # Two-tailed test

# Step 5: Define significance level
alpha = 0.05  # 95% confidence level
z_critical = norm.ppf(1 - alpha / 2)  # Z-critical value for 95% CI (¬±1.96)

# Step 6: Interpretation
if abs(z_score) > z_critical:
    print(f"Reject the null hypothesis (H‚ÇÄ). Z-score: {z_score:.2f}, P-value: {p_value:.4f}")
else:
    print(f"Fail to reject the null hypothesis (H‚ÇÄ). Z-score: {z_score:.2f}, P-value: {p_value:.4f}")

# Step 7: Visualization of Decision Regions
x_values = np.linspace(-3, 3, 1000)
y_values = norm.pdf(x_values, 0, 1)

plt.plot(x_values, y_values, color="blue", label="Standard Normal Distribution")

# Highlight rejection regions
plt.fill_between(x_values, y_values, where=(x_values < -z_critical) | (x_values > z_critical), color='red', alpha=0.5, label="Rejection Region")

# Mark calculated Z-score
plt.axvline(z_score, color="black", linestyle="dashed", label=f"Z-score: {z_score:.2f}")

plt.axvline(-z_critical, color="green", linestyle="dashed", label="Critical Value: -1.96")
plt.axvline(z_critical, color="green", linestyle="dashed", label="Critical Value: +1.96")

plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.title("Two-Tailed Z-test Decision Region")
plt.legend()
plt.grid()
plt.show()
```

5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_type1_type2_errors(mu_null=0, mu_alt=1, sigma=1, alpha=0.05):
    """
    Visualizes Type 1 and Type 2 errors on a normal distribution curve.

    Parameters:
    - mu_null: Mean under null hypothesis (H0)
    - mu_alt: Mean under alternative hypothesis (H1)
    - sigma: Standard deviation
    - alpha: Significance level (default: 0.05)
    """
    # Compute critical value for rejection of H0
    critical_value = norm.ppf(1 - alpha, loc=mu_null, scale=sigma)

    # Generate x-values
    x = np.linspace(mu_null - 4*sigma, mu_alt + 4*sigma, 1000)

    # Compute probability density functions (PDFs)
    y_null = norm.pdf(x, mu_null, sigma)  # H0 distribution
    y_alt = norm.pdf(x, mu_alt, sigma)    # H1 distribution

    # Plot the distributions
    plt.plot(x, y_null, color='blue', label="Null Distribution (H‚ÇÄ)")
    plt.plot(x, y_alt, color='green', label="Alternative Distribution (H‚ÇÅ)")

    # Shade Type 1 Error (alpha)
    plt.fill_between(x, y_null, where=(x > critical_value), color='red', alpha=0.5, label="Type 1 Error (Œ±)")

    # Shade Type 2 Error (beta)
    plt.fill_between(x, y_alt, where=(x < critical_value), color='purple', alpha=0.5, label="Type 2 Error (Œ≤)")

    # Add reference lines
    plt.axvline(critical_value, color='black', linestyle='dashed', label="Critical Value")

    plt.xlabel("Test Statistic")
    plt.ylabel("Probability Density")
    plt.title("Type 1 & Type 2 Errors in Hypothesis Testing")
    plt.legend()
    plt.grid()
    plt.show()

# Call function to plot Type 1 & Type 2 errors
plot_type1_type2_errors()
```

6.  Write a Python program to perform an independent T-test and interpret the results.

```python
import numpy as np
from scipy.stats import ttest_ind

# Step 1: Generate two independent sample datasets
np.random.seed(42)  # Ensure reproducibility
group_A = np.random.normal(loc=100, scale=15, size=30)  # Mean=100, Std=15
group_B = np.random.normal(loc=110, scale=15, size=30)  # Mean=110, Std=15

# Step 2: Perform independent T-test
t_stat, p_value = ttest_ind(group_A, group_B)

# Step 3: Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print(f"Reject the null hypothesis (H‚ÇÄ). T-statistic: {t_stat:.2f}, P-value: {p_value:.4f}")
else:
    print(f"Fail to reject the null hypothesis (H‚ÇÄ). T-statistic: {t_stat:.2f}, P-value: {p_value:.4f}")
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
import scipy.stats as stats

def calculate_confidence_interval(data, confidence=0.95):
    """
    Calculates the confidence interval for a sample mean.

    Parameters:
    - data: list or numpy array (sample data)
    - confidence: confidence level (default: 95%)

    Returns:
    - (lower_bound, upper_bound): Tuple representing the confidence interval
    """

    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Use ddof=1 for sample standard deviation
    sample_size = len(data)

    # Compute Z-score for confidence level
    z_score = stats.norm.ppf(1 - (1 - confidence) / 2)

    # Compute margin of error
    margin_of_error = z_score * (sample_std / np.sqrt(sample_size))

    # Compute confidence interval bounds
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return (lower_bound, upper_bound)

# Example usage:
data = np.random.normal(loc=100, scale=15, size=30)  # Generating sample data
ci = calculate_confidence_interval(data)

print(f"95% Confidence Interval: ({ci[0]:.2f}, {ci[1]:.2f})")
```

10.  Write a Python program to calculate the margin of error for a given confidence level using sample data.

```python
import numpy as np
import scipy.stats as stats

def calculate_margin_of_error(data, confidence=0.95):
    """
    Calculates the margin of error for a given confidence level using sample data.

    Parameters:
    - data: list or numpy array (sample data)
    - confidence: confidence level (default: 95%)

    Returns:
    - Margin of Error (MOE)
    """

    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Use ddof=1 for sample standard deviation
    sample_size = len(data)

    # Compute Z-score for confidence level
    z_score = stats.norm.ppf(1 - (1 - confidence) / 2)

    # Compute margin of error
    margin_of_error = z_score * (sample_std / np.sqrt(sample_size))

    return margin_of_error

# Example usage:
data = np.random.normal(loc=100, scale=15, size=30)  # Generate sample data
moe = calculate_margin_of_error(data)

print(f"Margin of Error for 95% confidence: {moe:.2f}")
```

11.  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
def bayes_theorem(prior, likelihood, marginal):
    """
    Computes the posterior probability using Bayes' Theorem.

    Parameters:
    - prior: P(A) - Prior probability
    - likelihood: P(B | A) - Probability of B given A is true
    - marginal: P(B) - Total probability of observing B

    Returns:
    - Posterior probability P(A | B)
    """
    return (likelihood * prior) / marginal

# Given Data
prior_prob = 0.02  # P(Disease)
likelihood = 0.90  # P(Positive Test | Disease)
false_positive_rate = 0.05  # P(Positive Test | No Disease)
no_disease_prob = 1 - prior_prob  # P(No Disease)

# Compute Marginal Probability P(Positive Test)
marginal_prob = (likelihood * prior_prob) + (false_positive_rate * no_disease_prob)

# Compute Posterior Probability P(Disease | Positive Test)
posterior_prob = bayes_theorem(prior_prob, likelihood, marginal_prob)

print(f"Probability of having disease given a positive test: {posterior_prob:.4f}")
```

12.  Perform a Chi-square test for independence between two categorical variables in Python.

```python
import numpy as np
import scipy.stats as stats

# Step 1: Create an observed frequency table
observed = np.array([[50, 30],  # Male choosing Product A & B
                     [20, 40]]) # Female choosing Product A & B

# Step 2: Perform Chi-square test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(observed)

# Step 3: Display results
print(f"Chi-square Statistic: {chi2_stat:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies:\n{expected}")

# Step 4: Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis (H‚ÇÄ): There is a significant association between Gender and Product Preference.")
else:
    print("Fail to reject the null hypothesis (H‚ÇÄ): Gender and Product Preference are independent.")
```

14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
